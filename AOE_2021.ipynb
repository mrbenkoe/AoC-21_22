```python
import pandas as pd
import numpy as np
```

```python
'''How many measurements are larger than the previous measurement?'''
```

```python
data = list(pd.read_csv('data/2021_01.txt',header=None)[0])
data
```

```python
len(data)
```

```python
first = np.array(data[:-1])
second = np.array(data[1:])

def counter_function(first, second):
    counter = 0
    for vi, val in enumerate(first):
        if second[vi] > val:
            counter += 1
    return counter

counter_function(first, second)  
```

```python
'''How many sums are larger than the previous sum?'''
```

```python
list_of_sums = []

for i in range(len(data)):
    if i+3 <= len(data):
        list_of_sums.append(sum(data[i:i+3]))
    
len(data) - len(list_of_sums) == 2
```

```python
first = np.array(list_of_sums[:-1])
second = np.array(list_of_sums[1:])

counter_function(first, second)  
```

# 02/1

```python
'''What do you get if you multiply your final horizontal position by your final depth?'''
```

```python
data = list(pd.read_csv('data/2021_02.txt',header=None)[0])
data
```

```python
len(data)
```

```python
forward = []
down = []
up = []

for val in data:
    value = int(val.split()[1])
    if 'forward' in val:
        forward.append(value)
    if 'down' in val:
        down.append(value)
    if 'up' in val:
        up.append(value)
```

```python
len(data) - len(forward) - len(down) - len(up)
```

```python
res = sum(forward) * (sum(down) - sum(up))
res
```

```python
'''
What do you get if you multiply your final horizontal position by your final depth?

a_new = a_old + DOWN
a_new = a_old - UP
    
h_new = h_old + FORWARD
d_new = d_old + a_old * FORWARD
    
'''
```

```python
data = list(pd.read_csv('data/2021_02.txt',header=None)[0])
data
```

```python
horizontal = 0
depth = 0
aim = 0

for val in data:
    value = int(val.split()[1])
    if 'forward' in val:
        horizontal += value
        depth = depth + aim * value
    if 'down' in val:
        aim += value
    if 'up' in val:
        aim -= value

print(horizontal, depth)
```
